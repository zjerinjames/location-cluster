# Cluster the coordinates into reuired cluster and plot it

```python
import pandas as pd
import numpy as np
import matplotlib.colors as colors
import folium

data = pd.read_csv('locations.csv')
data
```

```python
data = np.column_stack((data['lat'], data['lon']))
data = data.tolist()
data
```

```python
locations = data

# Create a map centered on the first coordinate
m = folium.Map(location=locations[0], zoom_start=12)

# Add a marker for each coordinate
for loc in locations:
    folium.Marker(location=loc).add_to(m)

# Display the map
m
```

```python
# # Generate a list of 25 different colors using the 'tab20' colormap
cmap = colors.ListedColormap(colors=['tab:blue', 'tab:orange', 'tab:green', 'tab:red', 'tab:purple', 'tab:brown', 'tab:pink', 'tab:gray', 'tab:olive', 'tab:cyan', 'lime', 'darkkhaki', 'darkslateblue', 'peru', 'darkorchid', 'aqua', 'blueviolet', 'burlywood', 'cadetblue', 'chocolate', 'coral', 'cornflowerblue', 'crimson', 'cyan', 'forestgreen', 'gold'])
colors = [colors.rgb2hex(cmap(i)) for i in range(cmap.N)]

# Create a dictionary mapping labels to colors
label_colors = dict(zip(range(25), colors))
```

```python
from sklearn.cluster import KMeans

# Define number of clusters required
n_clusters = 25

# Initialize the cluster centroids randomly
kmeans = KMeans(n_clusters=n_clusters, init='random', n_init=10)

# Perform the initial clustering
labels = kmeans.fit_predict(data)

# Define a color for each label
# label_color_dict = {0: 'red', 1: 'blue', 2: 'green', 3: 'purple', 4: 'orange'}

# Create a folium map
m = folium.Map(location=locations[0], zoom_start=5)

# Add a marker for each data point, colored according to its label
for i, point in enumerate(data):
    folium.CircleMarker(location=point, radius=5, fill_color=label_colors[labels[i]], color=label_colors[labels[i]], fill_opacity=1).add_to(m)

# Display the map
m
```

```python
# number of elements per cluster. The number of elements in each cluster wont be same usually
clusters, count =  np.unique(labels, return_counts=True)
print('Cluster No:', clusters)
print('No of elements:', count)
```
